```python
# Helper to Read Events from Event File

'''
Collection of code to interact with
EQ event lists.
'''

import datetime
import pandas as pd

class eq_events:
    def __init__(self,path_to_event_csv):
        self.data_frame = pd.read_csv(path_to_event_csv,sep='|')

    def add_column(self,col_name,init_val):
        self.data_frame[col_name] = init_val

    def get_date_time(self,row_num,dt_col_name='Time'):
        dt = self.data_frame[dt_col_name][row_num]
        dt_list = dt.split('T')
        date_list = dt_list[0].split('-')
        time_list = dt_list[1].split(':')

        dt_obj = datetime.datetime(int(date_list[0]),int(date_list[1]),int(date_list[2]),
                          int(time_list[0]),int(time_list[1]),int(time_list[2]))
        return dt_obj
```

```python
# Helper to Iterate Through Local Event DF and Return Search Param
def local_iterate(eq_event_object,minmag=2):
    '''
    Returns a list where each element is a row from the DF
    '''
    res = list()
    for row in eq_event_object.data_frame.itertuples():
        if row.Magnitude > minmag:
            res.append(row)
    return res

def get_search_param(nt):
    '''
    Takes in a named_tuple from list generated in
    local_iterate() and returns the search_param
    '''
    lat = nt.Latitude
    lon = nt.Longitude
    time = nt.Time
    mag = (nt.Magnitude,nt.MagType)
    return lat,lon,time,mag
```

```python
# Function to Check if a Station Was Active During an Event
def was_station_active(inventory,sta_name,time,channel):
    t1 = obspy.UTCDateTime(time) - (20*60)
    t2 = t1 + (40*60)
    inv_slice = (inventory.select(station=sta_name)).select(starttime=t1,endtime=t2)
    if len(inv_slice) == 0:
        return 0 # No network object
    else:
        length = len(inv_slice[0][0].select(channel=channel))
    return length
```

```python
# Data Structure to Capture Membership Info
class event_membership:
    def __init__(self,num_events,st_list,instr_list):
        self.row_count = num_events
        self.col_count = len(st_list)
        self.st_list = st_list
        self.instr_list = instr_list
        self.instr_count = len(instr_list)
        self.data = np.zeros((self.instr_count,self.row_count,self.col_count),dtype=int)
    
    def print_summary(self,instrument):
        print("Printing Summary for Channel {}".format(instrument))
        print("================================")
        for i in range(self.col_count):
            instr_index = self.instr_list.index(instrument)
            print("{} in {}/{} events".format(self.st_list[i],self.data[instr_index,:,i].sum(),self.row_count))
    
    def get_col_totals(self,instrument):
        res = list()
        for i in range(self.col_count):
            instr_index = self.instr_list.index(instrument)
            res.append(self.data[instr_index,:,i].sum())
        return res
    
    def plot_summary(self,instr_type='H',custom_str=""):
        fig,ax = plt.subplots()
        series_1 = self.get_col_totals('H{}E'.format(instr_type))
        series_2 = self.get_col_totals('H{}N'.format(instr_type))
        series_3 = self.get_col_totals('H{}Z'.format(instr_type))
        x = np.arange(len(self.st_list))
        ax.bar(x,series_1,width=0.25,label='H{}E'.format(instr_type))
        ax.bar(x+0.25,series_2,width=0.25,label='H{}N'.format(instr_type))
        ax.bar(x+0.5,series_3,width=0.25,label='H{}Z'.format(instr_type))
        ax.set_xticks(x+0.25)
        ax.set_xticklabels(self.st_list,rotation=45)
        for ind,val in enumerate(series_2):
            plt.text(x=x[ind]+0.25,y=val+2,s="{}".format(val),ha='center')
        plt.xlabel('Station Code')
        plt.ylabel('Number of Events')
        plt.legend()
        plt.grid()
        if instr_type == 'H':
            instr_str = "Seismometer"
        elif instr_type == 'N':
            instr_str = "Accelerometer"
        else:
            instr_type = "<Unsupported Instrument>"
        plt.title("Channel Data Presence | Instrument : {} | {}".format(instr_str,custom_str))
        fig.tight_layout()
        width = fig.get_figwidth()
        fig.set_figwidth(width+3)
    
    def get_value_at(self,instr,ev,stat):
        instr_index = self.instr_list.index(instr)
        st_index = self.st_list.index(stat)
        return self.data[instr_index,ev,st_index]
    
    def increment_value_at(self,instr,ev,stat,new_val):
        instr_index = self.instr_list.index(instr)
        st_index = self.st_list.index(stat)
        self.data[instr_index,ev,st_index] += new_val
    
    def update_membership_info(self,inventory,ev_list):
        # Loop Through Event List and Populate Membership Table    
        for index, ev in enumerate(tqdm(ev_list)):
            time = ev.Time
            for station in self.st_list:
                for instr in self.instr_list:
                    val = was_station_active(inventory,station,time,instr)
                    if(val>0):
                        self.increment_value_at(instr,index,station,val)
```
